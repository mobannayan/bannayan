```python
import matplotlib.pyplot as plt
import numpy as np

# Data
categories = ['Bad', 'Medium', 'High', 'Very High']
max_values = [49.156472, 49.397134, 48.824602, 15.290761]
min_values = [31.889680, 30.239993, 30.813710, 35.290761]
mean_values = [30.888685, 28.421316, 30.814923, 15.290761]

# Create the figure and axis
fig, ax = plt.subplots(figsize=(10, 6))

# Set the positions of the bars
y_indexes = np.arange(len(categories))

# Create bars for max, min, and mean values
# Plotting the max bar
max_bars = ax.barh(y_indexes, max_values, color='brown', edgecolor='black', label='Max')

# Plotting the min bar
min_bars = ax.barh(y_indexes, min_values, color='orange', edgecolor='black', label='Min')

# Plotting the mean bar
mean_bars = ax.barh(y_indexes, mean_values, color='green', edgecolor='black', label='Mean')

# Add arrows for mean values
for i in range(len(mean_values)):
    ax.annotate('', xy=(mean_values[i] + 1, i), xytext=(mean_values[i], i),
                arrowprops=dict(arrowstyle='->', color='green', lw=2))

# Set y-ticks and labels
ax.set_yticks(y_indexes)
ax.set_yticklabels(categories)

# Set x-ticks and limits
ax.set_xlim(0, 60)
ax.set_xticks(np.arange(0, 61, 5))

# Add labels and title
ax.set_xlabel('Values', fontsize=14)
ax.set_title('Bulk Density Analysis', fontsize=16, loc='left', pad=20)

# Add legend
ax.legend()

# Show the plot
plt.tight_layout()
plt.show()
```
