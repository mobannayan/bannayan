```python
myweather=pd.read_csv('F:\Climate corporation-Bayer-Ettain Group - Copy\practice_python\weather_2.csv')
myweather.head()
```

```python
plt.plot(myweather['Hum'], myweather['Tmax'])
plt.plot(myweather['Hum'], myweather['Tmin'])
plt.plot(myweather['Hum'], myweather['Rain '])
plt.show()
```

```python
subset = myweather[(myweather['t']>100) & (myweather['t']<=150)]

plt.plot(subset['Hum'], subset['Tmax'])
plt.plot(subset['Hum'], subset['Tmin'])
plt.plot(subset['Hum'], subset['Rain '])
plt.show()
```

```python
plt.scatter(subset['Hum'], subset['Tmax'])
plt.scatter(subset['Hum'], subset['Tmin'])
plt.scatter(subset['Hum'], subset['Rain '])
plt.show()
```

```python
plt.style.use('default')
plt.scatter(subset['Hum'], subset['Tmax'], label='clay')
plt.scatter(subset['Hum'], subset['Tmin'], label='sand')
plt.scatter(subset['Hum'], subset['Rain '], label='silt')
plt.grid()
plt.legend()
plt.show()
```
