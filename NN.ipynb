```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Our equation for data generation
x = np.arange(-80, 81, 0.5)
y = x**2 + 5*x +3
```

```python
# Compare the two modelsx.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
x_train.shape, x_test.shape
```

```python
model1 = keras.Sequential()
model1.add(keras.layers.Dense(units=1024, activation='relu', input_shape=[1]))
model1.add(keras.layers.Dense(units=1))
```

```python
model1.compile(optimizer='adam', loss="mean_squared_error")
```

```python
model1.summary()
```

```python
hist1 = model1.fit(x_train, y_train, batch_size=64, epochs=100, validation_data=(x_test, y_test))
```

```python
model2 = keras.Sequential()
model2.add(keras.layers.Dense(units=256, activation='relu', input_shape=[1]))
model2.add(keras.layers.Dense(units=256, activation='relu'))
model2.add(keras.layers.Dense(units=256, activation='relu'))
model2.add(keras.layers.Dense(units=256, activation='relu'))
model2.add(keras.layers.Dense(units=1))

```

```python
model2.compile(optimizer='adam', loss="mean_squared_error")
```

```python
model2.summary()
```

```python
model2.fit(x_train, y_train, batch_size=64, epochs=100, validation_data=(x_test, y_test))
```

```python
# Model comparison
y_pred1 = model1.predict(x_test)
y_pred2 = model2.predict(x_test)
```

```python
plt.scatter(x_test, y_test, color='red')
plt.plot(x_test, y_pred1  , color='blue')
plt.plot(x_test, y_pred2  , color='green')
plt.show()
```

```python
# Compare the two models
plt.plot(hist1.history['val_loss'])
plt.show()
```

```python
plt.plot(hist1.history['val_loss'][70:])
plt.show()
```

```python
optimizer = keras.optimizers.RMSprop(learning_rate=0.0001)
loss = keras.losses.mean_squared_error
metric = tf.metrics..RootMeanSquaredError()
model2.compile(optimizer=optimizer, loss=loss, metrics=metric)
```

```python
early_stop = keras.callbacks.EarlyStopping(monior='val_loss', pateince=10,restire_best_weights=True)
```

```python
model2.fit(x_train, y_train, epochs=100, batch_size=128, callbacks=[early_stop], validate_data=(x_test, y_test))
```
