```python
# List of names of cities
cities = ["Mashhad", "Shiraz", "Rasht", "Tabriz", "Isfahan",
          "Tehran", "Kerman", "Ahvaz", "Yazd", "Kermanshah"]

labels = []
for city in cities[:-1]: # Mdeans all cities except the last one (first 9 cities)
    labels.extend([city] * 365) # create 365 a list with the city name repated 365 times

# city >>> remaining rows(367)
labels.extend([cities[-1]] * (len(df100) - len(labels)))


# df100.insert(loc, column, value)
df100.insert(0, 'city', labels)


```
