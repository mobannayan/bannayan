Summary of slicing forms
sequence[start:stop:step]
start → index where slice begins (inclusive).
stop → index where slice ends (exclusive).
step → how much to move forward (or backward if negative).
[:] → copy all
[start:] → from start index to end
[:stop] → from beginning to stop index (exclusive)
[start:stop] → between two indices
[::step] → every nth element
[::-1] → reverse sequence
