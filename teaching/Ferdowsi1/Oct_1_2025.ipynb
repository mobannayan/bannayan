```python
# apply
import pandas as pd
s =pd.Series([100, 200, 300, 400])

def labeling(price):
    if price < 200:
        return "low"
    elif price <400:
        return "Medium"
    else:
        return 'Expensive'


print(s.apply(labeling))
```

```python
s = pd.Series([10, 20, 30])
new_s = s.apply(lambda x: x + 5)
print(new_s)
```
