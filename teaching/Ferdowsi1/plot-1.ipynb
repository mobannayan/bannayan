```python
hot_days = (m['Tmax']>20).sum()
cold_days = (m['Tmax']<10).sum()

mylist = [hot_days, cold_days]

colors = ['yellow', 'blue']

plt.figure(figsize=(5, 5))
plt.pie(mylist, labels=['Hot', 'Cold'], colors = colors, startangle=20)

plt.title('Temperature dist.')
plt.show()
```

```python
hot_days = (m['Tmax']>20).sum()
cold_days = (m['Tmax']<10).sum()
normal_days = ((m['Tave']>10) & (m['Tave']<15)).sum()
mylist = [hot_days, cold_days, normal_days]

colors = ['yellow', 'blue', 'Green']

plt.figure(figsize=(5, 5))
plt.pie(mylist, labels=['Hot', 'Cold', 'Normal'], colors = colors, startangle=20)

plt.title('Temperature dist.')
plt.show()
```
