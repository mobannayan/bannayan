start â†’ index where slice begins (inclusive).
stop â†’ index where slice ends (exclusive).
step â†’ how much to move forward (or backward if negative).
[:] â†’ copy all
[start:] â†’ from start index to end
[:stop] â†’ from beginning to stop index (exclusive)
[start:stop] â†’ between two indices
[::step] â†’ every nth element
[::-1] â†’ reverse sequence

```python
import pandas as pd
import random

# List of foods
foods = ["Apple", "Banana", "Carrot", "Cheese", "Bread", "Milk", "Egg", "Yogurt", "Chicken", "Salmon"]

# Nutritional range estimates per food item (average per 100g or typical serving)
nutrition_info = {
    "Apple":     {"Calories": (50, 60),   "Protein_g": (0.2, 0.5),  "Fat_g": (0.1, 0.3),   "Carbs_g": (13, 16),   "Sugar_g": (10, 12)},
    "Banana":    {"Calories": (85, 100),  "Protein_g": (1.0, 1.5),  "Fat_g": (0.2, 0.5),   "Carbs_g": (22, 27),   "Sugar_g": (12, 15)},
    "Carrot":    {"Calories": (35, 45),   "Protein_g": (0.5, 1.0),  "Fat_g": (0.1, 0.3),   "Carbs_g": (8, 10),    "Sugar_g": (4, 6)},
    "Cheese":    {"Calories": (300, 400), "Protein_g": (20, 25),    "Fat_g": (25, 35),     "Carbs_g": (1, 3),     "Sugar_g": (0, 1)},
    "Bread":     {"Calories": (250, 300), "Protein_g": (7, 9),      "Fat_g": (3, 5),       "Carbs_g": (45, 55),   "Sugar_g": (3, 5)},
    "Milk":      {"Calories": (90, 110),  "Protein_g": (3, 4),      "Fat_g": (3, 5),       "Carbs_g": (10, 12),   "Sugar_g": (9, 11)},
    "Egg":       {"Calories": (70, 90),   "Protein_g": (6, 8),      "Fat_g": (5, 7),       "Carbs_g": (0.5, 1.0), "Sugar_g": (0, 0.5)},
    "Yogurt":    {"Calories": (90, 120),  "Protein_g": (5, 10),     "Fat_g": (2, 4),       "Carbs_g": (10, 15),   "Sugar_g": (8, 12)},
    "Chicken":   {"Calories": (165, 200), "Protein_g": (25, 30),    "Fat_g": (3, 8),       "Carbs_g": (0, 1),     "Sugar_g": (0, 0)},
    "Salmon":    {"Calories": (200, 250), "Protein_g": (20, 25),    "Fat_g": (10, 15),     "Carbs_g": (0, 1),     "Sugar_g": (0, 0)},
}

# Generate 200 rows of food intake data
data = []
for day in range(1, 201):
    food = random.choice(foods)
    n = nutrition_info[food]
    row = {
        "Day": day,
        "Food": food,
        "Calories": round(random.uniform(*n["Calories"]), 1),
        "Protein_g": round(random.uniform(*n["Protein_g"]), 1),
        "Fat_g": round(random.uniform(*n["Fat_g"]), 1),
        "Carbs_g": round(random.uniform(*n["Carbs_g"]), 1),
        "Sugar_g": round(random.uniform(*n["Sugar_g"]), 1),
    }
    data.append(row)

# Create DataFrame
foods = pd.DataFrame(data)

# Display first 5 rows
foods.head()
```

```python
import pandas as pd

logs = [
  "2025-09-03 10:02:12 | SM:23.4 | TEMP:17.8",
  "2025-09-03 10:07:12 | SM:23.1 | TEMP:18.0",
  "2025-09-03 10:12:15 | SM:22.9 | TEMP:18.3"
]

data = [] # Create a empty list 
for log in logs: #Loop through strings in log
    parts = log.split("|") # removes th evertical symbol
    timestamp = parts[0].strip() # Take the first part (date + time). .strip() removes extra spaces.
                                  #"2025-09-03 10:02:12 " â†’ "2025-09-03 10:02:12"
    sm = float(parts[1].split(":")[1])
#Take the second part " SM:23.4 " >>> Split it by ":" â†’ [" SM", "23.4"] >>>>Pick index [1] â†’ "23.4"
# >>>> Convert it to a float â†’ 23.4


    temp = float(parts[2].split(":")[1]) # Same idea for temperature: " TEMP:17.8" â†’ 17.8
    
    data.append([timestamp, sm, temp])

df = pd.DataFrame(data, columns=["timestamp", "soil_moisture", "temperature"])
print(df)
```

```python
posts = [
    "Just tried plant-based chicken nuggets, surprisingly good!",
    "Energy drinks keep me alive during finals, but I know theyâ€™re unhealthy ðŸ˜…",
    "Fermented foods like kimchi and kefir are amazing for gut health!",
    "Too much junk food on campusâ€¦ we need better healthy snack options.",
    "Protein shakes are my go-to after gym sessions."
]
```

```python
# Preprocessing (Structuring the Data)
# We convert raw text into structured features using TF-IDF (text â†’ numbers).

import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(stop_words="english")
# What is TfidfVectorizer?# Comes from scikit-learn (from sklearn.feature_extraction.text import TfidfVectorizer).
# It converts unstructured text into a numerical matrix that a machine learning model can use.
# It uses TF-IDF (Term Frequency â€“ Inverse Document Frequency).

# TF-IDF Concept Term Frequency (TF): how often a word appears in a document.
# Inverse Document Frequency (IDF): reduces the weight of very common words that appear everywhere (like food, good, day).


# stop_words="english" mean?  >>> Stop words = very common words (e.g., the, is, in, and, of).These donâ€™t add 
# much meaning, so the vectorizer ignores them. Example: â€œThe food is very goodâ€ â†’ without stop words: â€œfood goodâ€.



X = vectorizer.fit_transform(posts)

df = pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names_out())
print(df.head())
```

```python
# 3. Analysis
# We can extract useful insights for Food Science students: a) Identify Trending Food Topics

word_counts = df.sum().sort_values(ascending=False)
print(word_counts.head(10))

```

```python
# b) Cluster Similar Food Habits
# Use unsupervised ML (KMeans clustering):

from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=2, random_state=42)
clusters = kmeans.fit_predict(X)

for i, post in enumerate(posts):
    print(f"Cluster {clusters[i]}: {post}")


# Example output:
# Cluster 0 â†’ â€œEnergy drinksâ€¦â€, â€œToo much junk foodâ€¦â€
# Cluster 1 â†’ â€œPlant-based nuggetsâ€¦â€, â€œFermented foodsâ€¦â€, â€œProtein shakesâ€¦â€
# So we see two groups:
# âž¡ï¸ Unhealthy/Convenience foods vs Healthy/Functional foods
```

```python
#glob.glob("data/*.csv") â†’ gets all CSV files in a folder.

import glob

# Get all CSV files in a folder (adjust path and pattern as needed)
files = glob.glob("data/*.csv")  # e.g., data/file1.csv, data/file2.csv

# Dictionary to store results
rain_counts = {}

for file in files:
    df100 = pd.read_csv(file)
    
    # Count days where Rain > 15
    count = (df100["Rain"] > 15).sum()
    
    # Store result, using file name as key
    rain_counts[file] = count

# Show results
for file, count in rain_counts.items():
    print(f"{file}: {count} days with Rain > 15")
```

```python
iles = glob.glob("data/*.csv")  # adjust folder path as needed

# Count rainy days
rain_counts = {}
for file in files:
    df100 = pd.read_csv(file)
    rain_counts[file] = (df100["Rain"] > 15).sum()

# Convert dictionary to DataFrame for easier plotting
rain_summary = pd.DataFrame(list(rain_counts.items()), columns=["File", "Days_Rain_gt_15"])

# Optional: shorten file names for better readability
rain_summary["File"] = rain_summary["File"].apply(lambda x: x.split("/")[-1])

# Plot bar chart
plt.figure(figsize=(10, 6))
plt.bar(rain_summary["File"], rain_summary["Days_Rain_gt_15"], color="skyblue")
plt.xlabel("File")
plt.ylabel("Number of Days with Rain > 15")
plt.title("Rainy Days (>15) Comparison Across Files")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
