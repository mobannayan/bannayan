```python
import random

# List of foods
foods = ["Apple", "Banana", "Carrot", "Cheese", "Bread", "Milk", "Egg", "Yogurt", "Chicken", "Salmon"]

# Nutritional range estimates per food item (average per 100g or typical serving)
nutrition_info = {
    "Apple":     {"Calories": (50, 60),   "Protein_g": (0.2, 0.5),  "Fat_g": (0.1, 0.3),   "Carbs_g": (13, 16),   "Sugar_g": (10, 12)},
    "Banana":    {"Calories": (85, 100),  "Protein_g": (1.0, 1.5),  "Fat_g": (0.2, 0.5),   "Carbs_g": (22, 27),   "Sugar_g": (12, 15)},
    "Carrot":    {"Calories": (35, 45),   "Protein_g": (0.5, 1.0),  "Fat_g": (0.1, 0.3),   "Carbs_g": (8, 10),    "Sugar_g": (4, 6)},
    "Cheese":    {"Calories": (300, 400), "Protein_g": (20, 25),    "Fat_g": (25, 35),     "Carbs_g": (1, 3),     "Sugar_g": (0, 1)},
    "Bread":     {"Calories": (250, 300), "Protein_g": (7, 9),      "Fat_g": (3, 5),       "Carbs_g": (45, 55),   "Sugar_g": (3, 5)},
    "Milk":      {"Calories": (90, 110),  "Protein_g": (3, 4),      "Fat_g": (3, 5),       "Carbs_g": (10, 12),   "Sugar_g": (9, 11)},
    "Egg":       {"Calories": (70, 90),   "Protein_g": (6, 8),      "Fat_g": (5, 7),       "Carbs_g": (0.5, 1.0), "Sugar_g": (0, 0.5)},
    "Yogurt":    {"Calories": (90, 120),  "Protein_g": (5, 10),     "Fat_g": (2, 4),       "Carbs_g": (10, 15),   "Sugar_g": (8, 12)},
    "Chicken":   {"Calories": (165, 200), "Protein_g": (25, 30),    "Fat_g": (3, 8),       "Carbs_g": (0, 1),     "Sugar_g": (0, 0)},
    "Salmon":    {"Calories": (200, 250), "Protein_g": (20, 25),    "Fat_g": (10, 15),     "Carbs_g": (0, 1),     "Sugar_g": (0, 0)},
}

# Generate 200 rows of food intake data
data = []
for day in range(1, 201):
    food = random.choice(foods)
    n = nutrition_info[food]
    row = {
        "Day": day,
        "Food": food,
        "Calories": round(random.uniform(*n["Calories"]), 1),
        "Protein_g": round(random.uniform(*n["Protein_g"]), 1),
        "Fat_g": round(random.uniform(*n["Fat_g"]), 1),
        "Carbs_g": round(random.uniform(*n["Carbs_g"]), 1),
        "Sugar_g": round(random.uniform(*n["Sugar_g"]), 1),
    }
    data.append(row)

# Create DataFrame
foods = pd.DataFrame(data)
```
